```python
Tamaño = input("Dame el tamaño del asteroide")
Velocidad = input("Dame la velocidad a la que se aproxima")

if(int(Tamaño) >= 25 and int(Velocidad) >= 25):
    print("Quien quiere ver un dinosaurio???")
elif(int(Tamaño) <= 25 and int(Velocidad) <= 20):
    print("No hay nada de que ver o preocuparse")
elif(int(Tamaño) <= 25 and int(Velocidad) >= 20):
    print("Veamos las luces en el cielo, pero no nos preocupemos")
elif(int(Tamaño) >= 25 and int(Velocidad) <= 20):
    print("preocumenos y sin espectaculo de luces")
```
