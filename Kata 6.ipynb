# Ejercicio 2 Kata 6

```python
# La lista

lista_planetas = ["Mercurio","Venus","Tierra","Marte","Jupiter","Saturno","Urano","Neptuno"]

# Se solicita un planeta

dame_un_planeta = input("Dime el nombre de un planeta, con la primera en mayuscula porfis")

# Search del planeta

busqueda = lista_planetas.index(dame_un_planeta)

# Planeta más cercano al sol

print("Este es el planeta más cercano " + dame_un_planeta)
print(lista_planetas[0:busqueda])

# Planeta más lejano al sol

print("Este es el planeta más lejano" + dame_un_planeta)
print(lista_planetas[busqueda + 1:])
```
