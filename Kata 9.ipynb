# Ejercicio 1 Kata 9 

```python
def reporte(tanque1, tanque2, tanque3):
    promedio = (tanque1 + tanque2 + tanque3) / 3
    return f"""Fuel Report:
    promedio total: {promedio}%
    1: {tanque1}%
    2: {tanque2}%
    3: {tanque3}% 
    """

print(reporte(80, 70, 85))

#Este si lo copie porque no le entendi muy bien 

def average(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items

print(average([80, 85, 81])) 


def generate_report(tanque1, tanque2, tanque3):
    return f"""Reporte final:
    Promedio total: {average([tanque1, tanque2, tanque3])}%
    Tercer tanque: {tanque1}%
    Segundo tanque: {tanque2}%
    Tercer tanque: {tanque3}% 
    """

print(generate_report(88, 76, 70))
```

```python
def reporte_de_mision(pre_lanzamiento, tiempo_de_vuelo, destino, segundo_tanque, primer_tanque):
    return f"""
    Mision hacia {destino}
    Distancia del viaje: {pre_lanzamiento + tiempo_de_vuelo} min
    Combustible total: {segundo_tanque + primer_tanque} galones
    """

print(reporte_de_mision(14, 51, "Moon", 200000, 300000))000000000
```

```python
def reporte_2(destino, *min, **reserva):
    return f"""
    Mison hacia {destino}
    Distancia del viaje: {sum(min)} minutes
    Reserva total: {sum(reserva.values())}
    """

print(reporte_2("Moon", 10, 15, 51, main=300000, external=200000))
```

```python
def reporte_3(destino, *min, **reserva):
    Reporte_total = f"""
    Mision hacia {destino}
    Distancia del viaje: {sum(min)} minutes
    Combustible total: {sum(reserva.values())}
    """
    for tanque, galones in reserva.items():
        Reporte_total += f"{tanque} tank --> {galones} gallons left\n"
    return Reporte_total

print(reporte_3("Moon", 8, 11, 55, main=300000, external=200000))
```
